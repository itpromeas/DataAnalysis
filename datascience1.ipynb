## Identify and handle missing values

### Convert "?" to NaN (Not a Number)

```python
import numpy as np
```

```python
# replace "?" to NaN
df2.replace("?", np.nan, inplace=True)
df2.head()
```

## Evaluate the missing data

```python
missingData = df2.isnull()
missingData.head()

# "false" stands for there is data and "true" stands for there is no data
```

```python
missing_data = df2.notnull()
missing_data.head()

# "true" stands for there is data and "false" stands for there is no data
```

## Count missing values in each column

```python
MissingDataListColumns = missing_data.columns.values.tolist()
for column in MissingDataListColumns:
    print(column)
    print(missing_data[column].value_counts())
```
